```python
# Ejercicio 1
numeros = (15, 8, 42, 3, 27, 19, 5)
numero_mayor = max(numeros)
numero_menor = min(numeros)
diff = numero_mayor - numero_menor

print(f"El numero mayor es: {numero_mayor}")
print(f"El numero menor es: {numero_menor}")
print(f"La diferencia es: {diff}")
```

```python
# Ejercicio 4
notas = (18,15,20,11,14,17,19)
acumulado = 0
for nota in notas:
    acumulado = nota + acumulado
prom = acumulado/len(notas)
maximo = max(notas)
minimo = min(notas)
print(f"El promedio de notas es de: {prom}")
print(f"La nota maxima es de: {maximo}")
print(f"La nota minima es de: {minimo}")   
```

```python
# Ejercicio 5
ventas=[('Lunes',120),('Martes',150),('Miércoles',95),('Jueves',180),('Viernes',210)]
dia_mayor, venta_mayor = max(ventas, key=lambda x: x[1])
total_ventas=sum(venta for dia, venta in ventas)
promedio = total_ventas/len(ventas)
print(f"El día con mayor venta fue el: {dia_mayor}")
print(f"El total de ventas fue de: {total_ventas}")
print(f"El promedio de ventas fue de: {promedio}")
```
