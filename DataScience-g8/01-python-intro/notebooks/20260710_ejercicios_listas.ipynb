```python
# Ejercicio 1
numeros = [4, 7, 2, 4, 9, 7, 1, 2, 10, 4]
numeros_distinct = []

for numero in numeros:
    if numero not in numeros_distinct:
        numeros_distinct.append(numero)

print (numeros_distinct)
```

```python
# Ejercicio 2
numeros = [12, 7, 9, 24, 31, 18, 5, 42, 13, 8]

numeros_pares = [
    numero
    for numero in numeros
    if numero % 2 == 0
]
numeros_impares = [
    numero
    for numero in numeros
    if numero % 2 != 0
]

print(f"numeros pares: {numeros_pares}")
print(f"numeros impares: {numeros_impares}")
```

```python
# Ejercicio 3
numeros = [45, 12, 89, 34, 89, 23, 67, 10]
numero_evaluado = 0
numero_evaluado2 =0
for numero in numeros:
    if numero >= numero_evaluado:
        numero_evaluado = numero

while numero_evaluado in numeros:
    numeros.remove(numero_evaluado)

for numero in numeros:
    if numero >= numero_evaluado2:
        numero_evaluado2 = numero

print (f"El segundo numero mayor es: {numero_evaluado2}")
```

```python
# Ejercicio 8

lista1 = [1, 3, 5, 7, 9, 11] 
lista2 = [2, 3, 6, 7, 10, 11]
lista_nueva = []

for numero in lista1:
    for numero2 in lista2:
        if numero == numero2:
            lista_nueva.append(numero)

print (f"Elementos comunes: {lista_nueva}")
```

```python
# Ejercicio 9

lista1 = [4, 8, 12, 16, 20] 
lista2 = [8, 10, 12, 18, 20]

lista_nueva = []
for numero in lista1:
    for numero2 in lista2:
        if numero == numero2:
            lista_nueva.append(numero)

for i in lista_nueva:
    lista1.remove(i)
for i in lista_nueva:
    lista2.remove(i)
lista1.extend(lista2)
print (f"Elementos comunes: {lista1}")
```
